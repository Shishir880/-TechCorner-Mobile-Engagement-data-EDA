```python

df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")


df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["Weekday"] = df["Date"].dt.day_name()


daily_sales = df.groupby("Date")["Sell Price"].sum()


monthly_sales = df.groupby("Month")["Sell Price"].sum()


weekday_sales = df.groupby("Weekday")["Sell Price"].sum()


fig, axes = plt.subplots(2, 2, figsize=(14, 10))


sns.lineplot(ax=axes[0, 0], x=daily_sales.index, y=daily_sales.values, marker="o", color="blue")
axes[0, 0].set_title("Daily Sales Trend")
axes[0, 0].set_xlabel("Date")
axes[0, 0].set_ylabel("Total Sales")


sns.barplot(ax=axes[0, 1], x=monthly_sales.index, y=monthly_sales.values, palette="viridis")
axes[0, 1].set_title("Monthly Sales Trend")
axes[0, 1].set_xlabel("Month")
axes[0, 1].set_ylabel("Total Sales")


sns.barplot(ax=axes[1, 0], x=weekday_sales.index, y=weekday_sales.values, palette="coolwarm")
axes[1, 0].set_title("Sales by Weekday")
axes[1, 0].set_xlabel("Weekday")
axes[1, 0].set_ylabel("Total Sales")


pivot_table = df.pivot_table(values="Sell Price", index="Day", columns="Month", aggfunc="sum")
sns.heatmap(pivot_table, cmap="Blues", ax=axes[1, 1], annot=True, fmt=".0f")
axes[1, 1].set_title("Sales Heatmap (Day vs. Month)")
axes[1, 1].set_xlabel("Month")
axes[1, 1].set_ylabel("Day")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Rolling Average Sales Trend (7-day moving average)
df["Rolling_Avg_Sales"] = daily_sales.rolling(window=7).mean()
# Use daily_sales.index for both x and y to ensure matching lengths
sns.lineplot(ax=axes[0, 0], x=daily_sales.index, y=daily_sales.rolling(window=7).mean(), marker="o", color="green")
axes[0, 0].set_title("7-Day Rolling Average Sales Trend")
axes[0, 0].set_xlabel("Date")
axes[0, 0].set_ylabel("Rolling Avg Sales")

# Distribution of Sales Transactions Per Day
transaction_counts = df.groupby("Date")["Cus.ID"].count()
sns.histplot(transaction_counts, bins=15, kde=True, color="purple", ax=axes[0, 1])
axes[0, 1].set_title("Daily Transaction Volume Distribution")
axes[0, 1].set_xlabel("Transactions Per Day")
axes[0, 1].set_ylabel("Frequency")

# Sales Boxplot by Month (to see variability)
sns.boxplot(ax=axes[1, 0], x=df["Month"], y=df["Sell Price"], palette="coolwarm")
axes[1, 0].set_title("Sales Price Distribution by Month")
axes[1, 0].set_xlabel("Month")
axes[1, 0].set_ylabel("Sell Price")

# Cumulative Sales Over Time
# Change: Use daily_sales.index for x and daily_sales.cumsum() for y
sns.lineplot(ax=axes[1, 1], x=daily_sales.index, y=daily_sales.cumsum(), marker="o", color="red")
axes[1, 1].set_title("Cumulative Sales Over Time")
axes[1, 1].set_xlabel("Date")
axes[1, 1].set_ylabel("Cumulative Sales")

# Adjust layout
plt.tight_layout()
plt.show()
```
